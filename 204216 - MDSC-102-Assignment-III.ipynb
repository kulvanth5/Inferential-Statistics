**Binomial Distribution**

```python
import matplotlib.pyplot as plt
import numpy as np
import math

def bi(n,x,p):
  return math.comb(n,x) * p**x * (1-p)**(n-x)

n = 10
x = np.arange(n+1)

f = [ bi(n,i,0.5) for i in x]

plt.bar(x,f)
plt.title('Binomial distribution')
plt.xlabel('values of x')
plt.ylabel('f(x)')
plt.show()
```

**Poisson Distribution**

```python
x = np.arange(0,20,1)

def p(lmd,x):
  return math.exp(-lmd) * lmd**x * math.factorial(x)**-1

f = [ p(5,i) for i in x ]

plt.bar(x,f)
plt.title('Poisson Distribution')
plt.xlabel('values of x')
plt.ylabel('f(x)')
plt.show()
```

**Exponential distribution**

```python
import random

lmd = 5

x = np.linspace(0,20,100)

def exp(lmd,x):
  return lmd * math.exp(-lmd*x)

f = [exp(lmd,i) for i in x]

plt.plot(x,f)
plt.title('Exponential Distribution')
plt.xlabel('values of x')
plt.ylabel('f(x)')
plt.show()
```

**Normal Distribution**

```python
m = 0
v = 1

def n(m,v,x):
  return math.sqrt(2*math.pi*v)**-1 * math.exp( -((x-m)**2) * (2*v)**-1 )

x = np.linspace(-5,5,100)

f = [ n(m,v,i) for i in x]

plt.plot(x,f)
plt.axis([-10,10,0,1])
plt.title('Normal Distribution')
plt.xlabel('values of x')
plt.ylabel('f(x)')
plt.show()
```

**Gamma Distribution**

```python
x = np.linspace(0,10,100)

def gm(al,bt,x):
  return (math.factorial(al-1)*bt**al)**-1 * x**(al-1) * math.exp(-x/bt)

f = [ gm(2,1,i) for i in x]

plt.plot(x,f)
plt.axis([0,10,0,1])
plt.title('Gamma Distribution')
plt.xlabel('values of x')
plt.ylabel('f(x)')
plt.show()
```
