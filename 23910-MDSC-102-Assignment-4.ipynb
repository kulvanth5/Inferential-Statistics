```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

features = ['class','Alcohol','Malic acid','Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline']

df = pd.read_csv('/content/wine.data')      #locally loaded the data into colab

df.columns = features

df
```

```python
sb.countplot(x='class',data=df)
```

```python
sb.scatterplot(x='Alcohol',y='Proline',hue='class',data=df)           #plotting alcohol vs proline
```

```python
fig , axes = plt.subplots(4,3,figsize=(20,20))    #setting the total size of the plotting
fig.subplots_adjust(hspace=0.5,wspace=0.5)        #adjusting the space b/w each subplot
fig.suptitle('Alcohol vs different attributes')   #title for the plotting

i = 0                                             #indexing for different subplots
j = 0

for feature in df.columns:                        # iterating over all the features

  if j > 2:
    j = 0
    i = i + 1

  if feature != 'Alcohol' and feature != 'class':                               #plotting all features w.r.t alcohol expect class
    sb.scatterplot(ax=axes[i,j],x='Alcohol',y=feature,hue='class',data=df)
    axes[i,j].set_title('Alcohol vs {}'.format(feature))                        #setting the title for each subplot
    j = j + 1
```
