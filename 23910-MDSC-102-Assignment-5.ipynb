```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

features = ['class','Alcohol','Malic acid','Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline']

df = pd.read_csv('/content/wine.data')      #locally loaded the data into colab

df.columns = features

df
```

```python
sb.countplot(x='class',data=df)
```

```python
sb.scatterplot(x='Alcohol',y='Proline',hue='class',data=df)           #plotting alcohol vs proline
```

```python
fig , axes = plt.subplots(4,3,figsize=(20,20))    #setting the total size of the plotting
fig.subplots_adjust(hspace=0.5,wspace=0.5)        #adjusting the space b/w each subplot
fig.suptitle('Alcohol vs different attributes')   #title for the plotting

i = 0                                             #indexing for different subplots
j = 0

for feature in df.columns:                        # iterating over all the features

  if j > 2:
    j = 0
    i = i + 1

  if feature != 'Alcohol' and feature != 'class':                               #plotting all features w.r.t alcohol expect class
    sb.scatterplot(ax=axes[i,j],x='Alcohol',y=feature,hue='class',data=df,palette=['blue','orange','green'])
    axes[i,j].set_title('Alcohol vs {}'.format(feature))                        #setting the title for each subplot
    j = j + 1
```

Plot 1:
Class 1 are more alcoholic and has more malic acid.

Plot 2:
Though class 2 are comparatively less alcoholic , they have more ash content.

Plot 3:
Class 1 has less alkalinity of ash than class 2 and class 3

Plot 4:
On an average , class 1 has more magnesium content than others.

Plot 5:
Class 3 has less number of phenols than class 1 and 2

Plot 6:
Clas 3 has less Flavanoids than the other 2 classes.

Plot 7:
Class 3 and 2 are spread over a wide range of non-flavanoid phenols where as class 1 are concentrated within 0.2-0.4 range of non-flavanoid phenols.

Plot 8:
Most of the class 3 samples are less than 1.5 in terms of proanthocyanins
class 1 range from 1.5 to 3 and class 2 are spread across more range.

Plot 9:
Color intensity of class 3 > class 1 > class 2

Plot 10:
Amount of Hue class 2 > class 1 > class 3

Plot 11:
OD280/OD385 of diluted wines in class 1 and class 2 is more than class 3

Plot 12:
As the Proline content increases , the wine becomes more alcoholic

```python
fig , axes = plt.subplots(4,3,figsize=(20,20))    #setting the total size of the plotting
fig.subplots_adjust(hspace=0.5,wspace=0.5)        #adjusting the space b/w each subplot
fig.suptitle('All contents w.r.t. to all classes.')
i = 0
j = 0

for feature in df.columns:

  if feature != 'Alcohol' and feature != 'class':
      sb.boxplot(data=df,ax=axes[i,j],x='class',y=feature,hue='class')
      j = j + 1

  if j > 2:
    j = 0
    i = i + 1
```

```python
fig , axes = plt.subplots(12,3,figsize=(30,60))
fig.subplots_adjust(hspace=0.5,wspace=0.5)

def hist(df,i):

  j = 0

  for col in df.columns:

    if col != 'class' and col != 'Alcohol':
      sb.histplot(ax=axes[i,j],kde=True,data=df,x=col)
      j = j + 1

    if j > 2:
      j = 0
      i = i + 1

hist(df[df['class']==1],0)
hist(df[df['class']==2],4)
hist(df[df['class']==3],8)
```
